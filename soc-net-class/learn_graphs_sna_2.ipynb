```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

### Graph generators

```python
# draw a complete graph
plt.figure(figsize = (10,7))
cg = nx.complete_graph(5, create_using = nx.DiGraph())
nx.draw(cg)
```

```python
print(cg.in_degree())
```

```python
print(cg.out_degree())
```

```python
# path graph
plt.figure(figsize = (10,7))
pg = nx.path_graph(5, create_using = nx.DiGraph())
nx.draw(pg)
```

```python
# gnr graph - like FB or Twitter
plt.figure(figsize = (10,7))
gnr = nx.gnr_graph(10, .5, create_using = nx.DiGraph())
nx.draw(gnr)
```

```python
print(gnr.in_degree())
```

```python
print(gnr.out_degree())
```

### complete bipartite graph

```python
plt.figure(figsize = (10,7))
cbg = nx.complete_bipartite_graph(20, 19)
nx.draw(cbg)
```

```python
cbg.order() #order is the same as the total number of nodes, 20 + 19
```

```python
cbg.size() # size is number of nodes multiplied, 20 x 19 
```

```python
# grid graph
gg = nx.grid_graph(dim = [11,7])
nx.draw(gg, node_size =300, node_color = 'firebrick')
```

```python
gg.order() #number of nodes multiplied
```

```python
gg.size() #number of edges
```

```python
gg.is_directed()
```

```python
import networkx as nx
%matplotlib inline
import matplotlib.pyplot as plt
```

### Section 15: Defining Functions

```python
def draw_graph():
    g = nx.Graph()
    g.add_edges_from([(1,2), (2,3), (1,3), (1,4)])
    nx.draw(g, with_labels = True, node_color = 'firebrick', node_size = 900)
```

```python
#draw graph from function
draw_graph()
```

```python
# create function for Barabasi_Albert
def barabasi_albert():
    h = nx.barabasi_albert_graph(100,10)
    hst = nx.degree_histogram(h)
    plt.hist(hst, bins = 20, color = 'green')
```

```python
barabasi_albert()
```

```python
#create nodes using a function
def simple_graph():
    j = nx.Graph()
    j.add_edges_from([('a','b'), ('b','c'), ('c', 'a')])
    nx.draw(j, with_labels = True, node_color = 'gold', node_size = 1000)
```

```python
simple_graph()
```
